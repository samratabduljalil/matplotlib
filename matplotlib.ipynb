```python
import matplotlib.pyplot as plt 
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
#scatter Plot 
X=np.random.rand(100)*100
Y=np.random.rand(100)*100

plt.figure(figsize=(10,7))
plt.scatter(X,Y,color='red',label='data')
plt.title('Scatter Plot',fontsize=26,color='red')
plt.suptitle('subtitle of scatter plot',fontsize=26)
plt.xlabel('weight',fontsize=16)
plt.ylabel('age' , fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```

```python
fruit_name =['apple','mango','banana','lichi']
weight=[100, 50, 120, 130]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode =(0.1,.2,0,0)

        

plt.figure(figsize=(6,6))
plt.pie(weight,labels=fruit_name,explode=explode,colors=colors)

plt.title('fruit pie',fontsize=16,color='red')
plt.suptitle('fruit weight pie',fontsize=20)
plt.legend()
plt.show
```

```python
fruit_name =['apple','mango','banana','lichi']
weight=[100, 50, 120, 130]

plt.figure(figsize=(6,6))
plt.bar(fruit_name,weight,color=['red', 'green', 'blue'])


plt.title('Bar Chart', fontsize=16)
plt.suptitle('Comparison of Categories', fontsize=10, color='gray')
plt.xlabel('Fruit name', fontsize=12)
plt.ylabel('weight', fontsize=12)
plt.grid(axis='y',linestyle='--',alpha=.4)
plt.show()
```

```python
data= np.random.randn(100)

plt.figure(figsize=(8, 6))

plt.hist(data,bins=10,color='purple',edgecolor='black')

plt.title('Histogram', fontsize=16)
plt.suptitle('Frequency Distribution', fontsize=10, color='gray')
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='dotted', alpha=0.7)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.figure(figsize=(8, 6))

plt.plot(x,y,marker='o',linestyle='--', label='Prime Numbers')
plt.title('Line Plot', fontsize=16)
plt.suptitle('Trend of Prime Numbers', fontsize=10, color='gray')
plt.xlabel('Index', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
X,Y =np.meshgrid(x,y)
Z=np.sin(np.sqrt(X**2 + Y**2))

fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(333,projection='3d')
surf= ax.plot_surface(X,Y,Z, cmap='viridis')

ax.set_title('3D Surface Plot', fontsize=16)
ax.set_xlabel('X-axis', fontsize=12)
ax.set_ylabel('Y-axis', fontsize=12)
ax.set_zlabel('Z-axis', fontsize=12)
```
