```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#scatter Plot 
X=np.random.rand(100)*100
Y=np.random.rand(100)*100

plt.figure(figsize=(10,7))
plt.scatter(X,Y,color='red',label='data')
plt.title('Scatter Plot',fontsize=26,color='red')
plt.suptitle('subtitle of scatter plot',fontsize=26)
plt.xlabel('weight',fontsize=16)
plt.ylabel('age' , fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```

```python
fruit_name =['apple','mango','banana','lichi']
weight=[100, 50, 120, 130]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode =(0.1,.2,0,0)

        

plt.figure(figsize=(6,6))
plt.pie(weight,labels=fruit_name,explode=explode,colors=colors)

plt.title('fruit pie',fontsize=16,color='red')
plt.suptitle('fruit weight pie',fontsize=20)
plt.show
```

```python
fruit_name =['apple','mango','banana','lichi']
weight=[100, 50, 120, 130]

plt.figure(figsize=(6,6))
plt.bar(fruit_name,weight,color=['red', 'green', 'blue'])


plt.title('Bar Chart', fontsize=16)
plt.suptitle('Comparison of Categories', fontsize=10, color='gray')
plt.xlabel('Fruit name', fontsize=12)
plt.ylabel('weight', fontsize=12)
plt.grid(axis='y',linestyle='--',alpha=.4)
plt.show()
```
